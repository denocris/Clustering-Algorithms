# Density Peaks

```python
import numpy as np
from pylab import *
```

```python
import matplotlib.cm as cmx
import matplotlib.colors as colors


def get_cmap(N):
    '''Returns a function that maps each index in 0, 1, ... N-1 to a distinct 
    RGB color.'''
    color_norm  = colors.Normalize(vmin=0, vmax=N-1)
    scalar_map = cmx.ScalarMappable(norm=color_norm, cmap='hsv') 
    def map_index_to_rgb_color(index):
        return scalar_map.to_rgba(index)
    return map_index_to_rgb_color

```

## Density Peaks: Aggregation Data Set

```python
f = open("DensityPeaks/Aggregation_Dataset/data/Aggregation.txt","r")

f = [line.strip() for line in f]
f = [line.strip().split('\t') for line in f ]
data = [[float(line[0]),float(line[1]),float(line[2])] for line in f ]
data_x = [line[0] for line in data ]
data_y = [line[1] for line in data ]
label = [line[2] for line in data ]

```

```python
def obs(k):
    obs_x = [line[0] for line in data if line[2] == k]
    obs_y = [line[1] for line in data if line[2] == k]
    return obs_x,obs_y

cmap = get_cmap(7)
for i in range(7):
    plt.scatter(obs(i + 1)[0], obs(i + 1)[1], c=cmap(i), s=7 ,alpha=0.5)

rcParams['figure.figsize'] = (10, 6)
plt.show()
```

```python
f = open("DensityPeaks/Aggregation_Dataset/data/decisiongraph_aggregation.txt","r")
f = [line.strip() for line in f]
f = [line.strip().split(' ') for line in f ]
f = [filter(None,line) for line in f ]

data = [[float(line[0]),float(line[1]),float(line[2]),float(line[3]),float(line[4])] for line in f ]
density = [line[3] for line in data ]
delta = [line[4] for line in data ]

plt.scatter(density, delta, c="blue", s=50 ,alpha=0.5)
plt.title("Decision Graph for Aggregation Data Set")
plt.xlabel('density')
plt.ylabel('delta')
plt.show()
```

The number of clusters find out by the density peaks algorithm is 7, according to the expectation!
